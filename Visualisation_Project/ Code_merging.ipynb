# Cleaning smoking data

```python
import pandas as pd
from pathlib import Path
```

```python
#create path to csv file:

smoking= Path("Resources/Raw/Raw_smoking.csv")
```

```python
smoking_df=pd.read_csv(smoking)
```

```python
#view smoking dataframe 
smoking_df
```

```python
#drop null values from smoking_df and reset index
smoking_df = smoking_df.dropna()
smoking_df=smoking_df.reset_index(drop=True)
smoking_df.head(10)
```

```python
# Rename 'Country' column to 'Country Name' in smoking_df
smoking_df.rename(columns={'Country': 'Country Name', 'Data.Daily cigarettes': 'Average cigarettes smoked per day by smoker',
                          'Data.Percentage.Male' : 'Percentage of smokers in male population', 'Data.Percentage.Female' : 'Percentage of smokers in female population',
                         'Data.Percentage.Total' : 'Percentage of smokers in population', 'Data.Smokers.Total' : 'Total number of smokers', 'Data.Smokers.Female': 'Total number of female smokers',
                        'Data.Smokers.Male' : 'Total number of male smokers'}, inplace=True)
smoking_df.head(10)
```

```python
# Write the new data to a CSV file
output_file_path= Path('Resources\smoking.csv') 
smoking_df.to_csv(output_file_path, index=False)
```

# Merge smoking data with GDP per capita data

```python
#create path to socioeconomic csv files:
gdp_path= Path('Resources/gdp_per_capita.csv')
unemployment_path= Path('Resources/unemployment.csv')
health_expenditure_path= Path('Resources\health_expenditure.csv')

#read csv files: 

gdp_df=pd.read_csv(gdp_path)
unemployment_df=pd.read_csv(unemployment_path)
health_expenditure_df=pd.read_csv(health_expenditure_path)
```

```python
#check GDP file
gdp_df.head()
```

```python
# Merge the smoking and GDP DataFrames
smoking_gdp = smoking_df.merge(gdp_df, on=['Country Name', 'Year'], how='left')
smoking_gdp.head()
```

```python
# Drop the 'Country Code' column from the DataFrame
smoking_gdp.drop(columns=['Country Code'], inplace=True)
smoking_gdp.head()
```

```python
#check unemployment dataframe
unemployment_df.head()
```

```python
# merge unemployment with smoking_gdp and drop "country code"
smoking_gdp_unemployment = smoking_gdp.merge(unemployment_df, on=['Country Name', 'Year'], how='left')

smoking_gdp_unemployment.drop(columns=['Country Code'], inplace=True)

smoking_gdp_unemployment
```

```python
#merge health expenditure with smoking_gdp_unemployment and drop "country code"
merge_all = smoking_gdp_unemployment.merge(health_expenditure_df, on=['Country Name', 'Year'], how='left')

merge_all.drop(columns=['Country Code'], inplace=True)

merge_all
```

```python
# set index as id column 
merge_all['Country_id'] = merge_all.index
merge_all.head()
```

```python
#reorder columns
cols = merge_all.columns.tolist()
cols = cols[-1:] + cols[:-1]
cols
```

```python
#reorder dataframe 
merge_all = merge_all[cols]
merge_all
```

```python
#Export to csv file namely Master_raw
output_file_path= Path('Resources\combined_data.csv') 
merge_all.to_csv(output_file_path, index=False)
```

```python
#Drop missing value in merged file
master = merge_all.dropna()
master
```

```python
#Export to csv file namely Master_cleaned
output_file_path= Path('Resources\Merged_data_null_dropped.csv') 
master.to_csv(output_file_path, index=False)
```
