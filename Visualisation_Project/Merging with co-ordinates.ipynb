```python
import pandas as pd
from pathlib import Path
import json
```

```python
#Load the .json file
json_borders = r"C:\Users\tamik\Documents\Repositories\Visualisation_Project\worldborders.json"
borders_df = pd.read_json(json_borders)
borders_df.head()
```

```python
# Extract and normalize the 'properties' column into separate columns for merging 
properties_df = pd.json_normalize(borders_df['features'].apply(lambda x: x['properties']))
geometries_df = pd.json_normalize(borders_df['features'].apply(lambda x: x['geometry']))
borders_normalised_df = pd.concat([properties_df, geometries_df], axis=1)
borders_normalised_df.head()
```

```python
borders_normalised_df.rename(columns={'name': 'Country Name'}, inplace=True)
borders_normalised_df.head()
```

```python
#create path to csv file:
combined = Path(r"C:\Users\tamik\Documents\Repositories\Visualisation_Project\Resources\combined_data.csv")
combined_df=pd.read_csv(combined)
combined_df.head()
```

```python
#create path to csv file:
smoking = Path(r"C:\Users\tamik\Documents\Repositories\Visualisation_Project\Resources\smoking.csv")
smoking_df = pd.read_csv(smoking)
smoking_df.head()
```

```python
#filter data to 2000 for one df and 2012 for another 
smoking_2000 = smoking_df[smoking_df['Year'] == 2000]
smoking_2000.head()

smoking_2012 = smoking_df[smoking_df['Year'] == 2012]
smoking_2012.head()
```

```python
# Merge the cordinates and the smoking DataFrames
coordinates_2000 = smoking_2000.merge(borders_normalised_df, on=['Country Name'], how='left')
coordinates_2000
coordinates_2000.drop_duplicates(subset=['Country Name'])
coordinates_2000
```

```python
#Drop missing value in merged file
cleaned_2000 = coordinates_2000.dropna()
cleaned_2000
cleaned_2000.drop(columns=['type'], inplace=True)
cleaned_2000.head()
```

```python
#remove duplicated somoa value
display(cleaned_2000.iloc[135])
cleaned_2000.drop(135, inplace=True)
cleaned_2000
```

```python
# Merge the cordinates and the smoking DataFrames
coordinates_2012 = smoking_2012.merge(borders_normalised_df, on=['Country Name'], how='left')
coordinates_2012
coordinates_2012.drop_duplicates(subset=['Country Name'])
coordinates_2012
```

```python
#Drop missing value in merged file
cleaned_2012 = coordinates_2012.dropna()
cleaned_2012
cleaned_2012.drop(columns=['type'], inplace=True)
cleaned_2012
```

```python
#remove duplicated somoa row
display(cleaned_2012.iloc[135])
cleaned_2012.drop(135, inplace=True)
cleaned_2012
```

```python
#export files to csv
output_file_path_2000= Path('Resources\coordinates_2000.csv') 
coordinates_2000.to_csv(output_file_path_2000, index=False)
```
